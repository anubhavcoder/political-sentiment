```python
def clean_tweet(tweet):
    return tweet.lower()
#     tweet = tweet.lower()
#     return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split()) 
```

```python
def get_tweets_list(account):
    tweets = []
    with open(f'./data/{account}.csv') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        keys = next(csv_reader)
        for row in csv_reader:
            tweet = {}
            tweet[keys[0]] = row[0]
            tweet[keys[1]] = row[1]
            tweet[keys[6]] = clean_tweet(row[6])
            tweets.append(tweet)
    return tweets
```

```python
# verify start and end data for all accounts
for acc in leaders_accounts:
    tl = get_tweets_list(acc)
    s = tl[0]['date']
    e = tl[-1]['date']
    assert s[0:4] == '2019'
    assert e[0:4] == '2015'
    print(f'acc: {acc}, start: {s}, end: {e}')
```

```python
def generate_wordcloud(tweets):
    tweets_con = ''
    for t in tweets:
        tweets_con += t['text']
    return WordCloud().generate(tweets_con)

def plot_wordcloud(cloud, leader, year):
    plt.figure(
    figsize=(6, 2),
    dpi=200)
    plt.title(f'Word Cloud of {leader} tweets in {year}')

    plt.imshow(cloud, interpolation='bilinear')
    plt.show()
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')

sid = SentimentIntensityAnalyzer()
```

### Define Analysis Methods

```python
def sentiment_over_time(tweets, sent):
    sentiment = []
    time = []
    for tweet in tweets:
        ss = sid.polarity_scores(tweet['text'])
        sentiment.append(ss[sent])
        date = tweet['date'].split('-')
        time.append(int(date[0]) + int(date[1]) + int(date[2].split(' ')[0]))
    return sentiment, time

def length_over_time(tweets):
    lengths = []
    time = []
    for tweet in tweets:
        lengths.append(len(tweet))
        date = tweet['date'].split('-')
        time.append(int(date[0]) + int(date[1]) + int(date[2].split(' ')[0]))
    return lengths, time


def plot_length_over_time(lengths, t, title):
    plt.figure(figsize=(12, 9))
    plt.title(title)
    plt.xticks(fontsize=14)
    plt.xticks(np.arange(min(t), max(t)+1, 25))
    plt.xlabel("Day of Year", fontsize=16)  
    plt.ylabel("Tweet Length", fontsize=16)  
    plt.plot(lengths, color="#3F5D7D")
    plt.show()

def plot_sentiment_over_time(s, t, title):
    plt.figure(figsize=(12, 9))
    plt.title(title)
    plt.xticks(fontsize=14)
    plt.xticks(np.arange(min(t), max(t)+1, 25))
    plt.xlabel("Day of Year", fontsize=16)  
    plt.ylabel("Sentiment Score", fontsize=16)  
    plt.plot(s, color="#3F5D7D")
    plt.show()
    
def get_tweets_with_keyword(tweets, keyword):
    new_tweets = [tweet for tweet in tweets if keyword in tweet['text']]
    return new_tweets
```
